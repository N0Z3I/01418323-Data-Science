```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dic =  {"day":['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],
        "temp":[24,25,29,35,23,20,23]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,ls='-.')
```

```python
plt.plot(x,y,color = (1,0.5,0))
```

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='s',ms=10,mec='black',mfc='purple')
```

```python
plt.plot(x,y,'s:r',lw=3,mec='green')
```

```python
plt.plot(x,y,'s:r',lw=3,mec='green')
plt.title('Temperature around this week')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.xticks(rotation=90)
plt.yticks([15,20,25,30,35,40])
plt.show()
```

```python
dic =  {"day":['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],
        "temp1":[24,25,29,35,23,20,23],
        "temp2":[10,14,9,12,11,15,17]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp1']
z = df['temp2']
```

```python
plt.plot(df['day'],df['temp1'],'s-.r',label='Temp1')
plt.plot(df['day'],df['temp2'],'o:g',label='Temp2')
plt.legend(loc='upper left')
```

```python
fig , axes = plt.subplots( nrows=2,ncols=2)
axes[0,0].plot(df['day'],df['temp1'],'s-.r')
axes[0,1].plot(df['temp1'],df['day'],'o:y')
axes[1,0].plot(df['temp1'],df['day'],'*-k')
axes[1,1].plot(df['day'],df['temp1'],'s--g')
axes[0,0].set_title('week1')
axes[0,1].set_title('week2')
axes[1,0].set_title('week3')
axes[1,1].set_title('week4')
fig.tight_layout()
```
