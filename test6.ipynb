```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dic =  {"day":['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],
        "temp":[24,25,29,35,23,20,23]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,ls='-.')
```

```python
plt.plot(x,y,color = (1,0.5,0))
```

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='s',ms=10,mec='black',mfc='purple')
```

```python
plt.plot(x,y,'s:r',lw=3,mec='green')
```

```python
plt.plot(x,y,'s:r',lw=3,mec='green')
plt.title('Temperature around this week')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.xticks(rotation=90)
plt.yticks([15,20,25,30,35,40])
plt.show()
```
