```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_excel('customer.xlsx')
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
att = df[['salary','purchased']]
label = df['gender']

att_train , att_test , class_train , class_test  = train_test_split(att,label,random_state=0,train_size=2)


scaler = StandardScaler()
scaler.fit(att_train)

att_train[['salary','purchased']] = scaler.transform(att_train)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(att_test,class_test)


model.score( scaler.transform(att_train) , class_train)
```

```python
result = pd.concat([att_test,class_test] , axis= 1)
result['predict'] = model.predict(att_test)
result
```
