## Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.arange(0,100)
y = x*2
z = x**2
df = pd.read_csv('Superstore.csv',encoding = 'iso-8859-1')
```

## Exercise 1


```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
```

## Exercise 2

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2 = fig.add_axes([0.15,0.7,0.3,0.2])
axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```

## Exercise 3



**ใช้  arrays x, y และ z เพื่อทำการ plot บนแกนที่สร้างจากข้อที่แล้ว  (Notice อย่าลืมกำหนด  x - limits และ y - limits )**

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2 = fig.add_axes([0.2,0.6,0.3,0.3])
axes2.plot(x,y,y,color="#7AAED2")
axes2.set_ylim(30,50)
axes2.set_xlim(20,22)
axes2.set_title('Zoom')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```

## Exercise 4

**จงใช้คำสั่ง plt.subplots(nrows=1, ncols=2)**

**จากนั้นให้ทำการ plot (x,y) และ plot (x,z) บนแกน axes  และให้ใช้งานคำสั่ง linewidth and style เพื่อตกแต่งเส้นของกราฟ**

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,0.5,0.5])
axes1.plot(x,y,color='b',ls='--',lw=3)

axes2 = fig.add_axes([0.6,0,0.5,0.5])
axes2.plot(x,z,color='r',lw=3)
```

## จงแสดงกราฟรายได้ของทุกปี

```python
df1 = df.groupby('Year')['Profit'].sum()
a = df1.index
b = df1
```

```python
plt.plot(a,b,color='black',ls='--',marker='o',mfc='red',mec='red')
plt.title('Profit')
plt.xticks(np.arange(2014,2018,1))
plt.show()
```

#### โค้ดต่อไปนี้ใช้ในสองข้อสุดท้าย

```python
df['Category'].unique()
```

```python
df[ df['Category'] == 'Furniture' ].groupby('Year').sum()['Profit']
```

##### Dictionary of Category

```python
arr_df = {}
for i in range(0,df['Category'].nunique()) :
    arr_df[df['Category'].unique()[i]] = df[ df['Category'] == df['Category'].unique()[i] ].groupby('Year').sum()['Profit']
```

```python
x = arr_df['Furniture'].index
y= arr_df['Furniture']
```

## จงแสดงกราฟรายได้ของแต่ละ Category ในแต่ละปีในกราฟเดียว 

```python
plt.plot(a,arr_df["Furniture"],"s-g",marker='^',color='#1A78FF',label='Furniture')
plt.plot(a,arr_df["Office Supplies"],"s-g",marker='o',color='#FF8F09',label='Office Supplies')
plt.plot(a,arr_df["Technology"],"s-g",label='Technology')
plt.legend(loc='upper left')
plt.xticks(np.arange(2014,2018,1))
plt.show()
```
