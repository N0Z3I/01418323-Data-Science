```python
g = sns.JointGrid(data=tips, x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.boxplot)
```

```python
g = sns.PairGrid(data=tips,hue='sex')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.boxplot)
```
