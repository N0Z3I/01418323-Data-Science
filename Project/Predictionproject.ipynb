```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('GlobalYouTubeStatistics.csv',encoding = 'iso-8859-1')
```

```python
df.info()
```

```python
df
```

```python
plt.figure(figsize=[4,3])
sns.scatterplot(data = df,x= 'subscribers',y= 'video views')
```

```python
df2 = df [['subscribers','video views']].dropna()
df2.head()
```

```python
from sklearn.cluster import  KMeans
```

```python
model =  KMeans(n_clusters=4,random_state=0)
model.fit(df2)
```

```python
plt.figure(figsize=[4,3])
sns.scatterplot(data = df2,x= 'subscribers',y= 'video views',hue = model.labels_ , palette='Set1')
plt.scatter( model.cluster_centers_[:,0], model.cluster_centers_[:,1],color='k', marker= '*')
```
